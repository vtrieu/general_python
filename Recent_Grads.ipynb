# Visualizing Earnings Based on College Major

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

recent_grads = pd.read_csv("recent-grads.csv") #reading in the recent-grads.csv data from FiveThirtyEight

```

```python
recent_grads = recent_grads.dropna() #drop rows containing missing values and reassign to recent_grads dataframe

cleaned_data_count = recent_grads.shape[0] #look up the number of rows in recent_grads after dropping rows with missing values
cleaned_data_count
```

## Relationships via Scatterplots

```python
# Sample Size vs. Median Income
recent_grads.plot(x='Sample_size', y='Median', kind='scatter', title='Sample Size vs. Median Income', figsize=(5,5))
```

When looking at the sample size for each major in relation to median income, we can see that the more popular majors, i.e. majors that have a higher sample size, make about the same amount of money as the less popular majors. We also see that for the least popular majors (which comprise the bulk of the data set), the median income is clustered between the 20k and 40k range. 

```python
# Sample Size vs. Unemployment Rate
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', title='Sample Size vs. Unemployment Rate', figsize=(8,5))
```

The data shows that less popular majors have a similar rate of unemployment as more popular majors.

```python
# Full Time vs. Median Salary
recent_grads.plot(x='Full_time', y='Median', kind='scatter', title='Full Time vs. Median Salary', figsize=(8,5))
```

There are instances where majors with lesser full time employees have a higher median salary.

```python
# Share of Women vs. Unemployment Rate
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter', title='Share of Women vs. Unemployment Rate', figsize=(5,5))
```

There is no relationship between a higher share of female employees with higher unemployment rate. 

```python
# Male Graduates vs. Median Income
recent_grads.plot(x='Men', y='Median', kind='scatter', title='Male Graduates vs. Median Income', figsize=(7,5))
```

```python
# Female Graduates vs. Median Income
recent_grads.plot(x='Women', y='Median', kind='scatter', title='Female Graduates vs. Median Income', figsize=(7,5))
```

```python
# Share of Women vs. Median Income
recent_grads.plot(x='ShareWomen', y='Median', kind='scatter', title='Share of Women vs. Median Income', figsize=(7,5))
```

Majors that were majority female made less money.
